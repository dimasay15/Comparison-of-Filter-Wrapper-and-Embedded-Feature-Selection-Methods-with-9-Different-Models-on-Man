```python
#tentukan metode scoring yang digunakan
scoring_rfe = {'acc': 'accuracy',
               'prec_micro': 'precision_micro',
               'rec_micro': 'recall_micro'}

#tentukan total fitur dan trees yang digunakan dalam proses klasifikasi ini
# n_feat = [10, 20, 30]
n_trees = [100, 150, 200]


for ntrees in n_trees:        
    #ambil n fitur input hasil seleksi fitur Fisher
    x_train_selected = x_train[max([list_anova, list_mi, list_fisher, list_reliefF, list_sfs, list_backward, list_rfe, list_Lasso, list_ElasticNet])]
    x_test_selected = x_test[max([list_anova, list_mi, list_fisher, list_reliefF, list_sfs, list_backward, list_rfe, list_Lasso, list_ElasticNet])]

    #Create a Gaussian Classifier
    clf_rfe = RandomForestClassifier(n_estimators=ntrees)

    #Train the model using the training sets
    clf_rfe.fit(x_train_selected, y_train)
    y_pred_rfe=clf_rfe.predict(x_test_selected)

    #hitung score model dari data train
    scores_rfe = cross_validate(clf_rfe, x_train_selected, y_train, scoring=scoring_rfe, cv=cv, return_train_score=True)

    print("Akurasi model RFE data Train dengan 30" + " fitur yang paling banyak muncul dari semua seleksi fitur dan " 
          + str(ntrees) + " trees:" + str(round(scores_rfe['train_acc'].mean(), 2)))
    print("Akurasi model RFE data Test dengan 30" + " fitur yang paling banyak muncul dari semua seleksi fitur dan " 
          + str(ntrees) + " trees:" + str(round(metrics.accuracy_score(y_test, y_pred_rfe), 2)))
    print("Precision model RFE data Train dengan 30" + " fitur yang paling banyak muncul dari semua seleksi fitur dan " 
          + str(ntrees) + " trees:" + str(round(scores_rfe['train_prec_micro'].mean(), 2)))
    print("Precision model RFE data Test dengan 30" + " fitur yang paling banyak muncul dari semua seleksi fitur dan " 
          + str(ntrees) + " trees:" + str(round(metrics.precision_score(y_test, y_pred_rfe, average='micro'), 2)))
    print("Recall model RFE data Train dengan 30" + " fitur yang paling banyak muncul dari semua seleksi fitur dan " 
          + str(ntrees) + " trees:" + str(round(scores_rfe['train_rec_micro'].mean(), 2)))
    print("Recall model RFE data Test dengan 30" + " fitur yang paling banyak muncul dari semua seleksi fitur dan " 
          + str(ntrees) + " trees:" + str(round(metrics.recall_score(y_test, y_pred_rfe, average='micro'), 2)))
    print(" ")
```
